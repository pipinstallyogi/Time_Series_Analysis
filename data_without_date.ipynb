```python
import pandas as pd
import numpy as np
df = pd.read_csv('TSLA_Without_Date.csv')
df.head()
```

```python
#date_range function in pandas will assign dates using the given parameters. B stands for Business day 
possible_dates = pd.date_range(start='2018-10-01', end='2018-10-31', freq = 'B')
possible_dates
```

```python
#setting dates as the column for our data
df.set_index(possible_dates, inplace = True)
```

```python
#If we also want to see the prices on weekends as well, which will obviously be similar to the friday closing prices
# https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#offset-aliases
df.asfreq('D', method ='pad')
```
